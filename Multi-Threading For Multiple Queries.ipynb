```python
import pandas as pd
import threading
import time

num_files=12

result2=[]
def data_parser(filename):
    try:
        df=pd.read_json(filename)
        df['Dogum Tarihi'] = pd.to_datetime(df['Dogum Tarihi'])
        kadikoy_2000=df[(df['Ilce']=='KADIKÖY') & (df['Dogum Tarihi']>'1990-01-01')]
        oz_kadikoy_2000=kadikoy_2000.loc[kadikoy_2000['soyisim'].str.contains('ÖZ')]
        data=oz_kadikoy_2000.to_dict(orient='records')
        for kisi in data:
            result2.append(kisi)
    except Exception as e:
        print(f'istisna var: {filename}')


start = time.time()

databases=[]
for i in range(1,(num_files+1)):
    filename=f'contactlist2_{i}.json'
    databases.append(filename)


threads=[]
for db in databases:
    th = threading.Thread(target=data_parser, args=(db,))
    threads.append(th)
    
for th in threads:
    th.start()
    
for th in threads:
    th.join()
    
end = time.time()
print(f'Current Execution Time:{end-start}')
print(len(result2))
```

```python
import pandas as pd
import time
import concurrent.futures


num_files=12

result3=[]
def data_parser(filename):
    try:
        df=pd.read_json(filename)
        df['Dogum Tarihi'] = pd.to_datetime(df['Dogum Tarihi'])
        kadikoy_2000=df[(df['Ilce']=='KADIKÖY') & (df['Dogum Tarihi']>'1990-01-01')]
        oz_kadikoy_2000=kadikoy_2000.loc[kadikoy_2000['soyisim'].str.contains('ÖZ')]
        data=oz_kadikoy_2000.to_dict(orient='records')
        for kisi in data:
            result3.append(kisi)
    except Exception as e:
        print(f'istisna var: {filename}')


start = time.time()

databases=[]
for i in range(1,(num_files+1)):
    filename=f'contactlist2_{i}.json'
    databases.append(filename)
    
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(data_parser, db) for db in databases]
    for future in concurrent.futures.as_completed(futures):
        future.result()  # To raise any exceptions occurred during the thread execution

end = time.time()
print(f'Current Execution Time: {end - start}')
```
